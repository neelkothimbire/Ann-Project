```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(1-x))

def relu(x):
    return np.maximum(0,x)

def tanh(x):
    return np.tanh(x)

def softmax(x):
    e_x=np.exp(x-np.max(x))
    return e_x/e_x.sum()
```

```python
x=np.linspace(-5,5,100)
y_sigmoid=sigmoid(x)
y_relu=relu(x)
y_tanh=tanh(x)
y_softmax=softmax(x)
plt.figure(figsize=(10,6))
```

```python

plt.subplot(2,2,1)
plt.plot(x,y_sigmoid,'b--',label='sigmoid')
plt.title('Sigmoid Activation')
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y_relu,'g--',label='relu')
plt.title('Relu Activation')
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y_tanh,'r--',label='tanh')
plt.title('tanh Activation')
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y_softmax,'y--',label='softmax')
plt.title('softmax Activation')
plt.xlabel('x')
plt.ylabel('f(x)')
```
